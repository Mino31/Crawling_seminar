# 크롤링 기본 맛보기

## 필요한 라이브러리 import

```python
from selenium import webdriver
from bs4 import BeautifulSoup
from wordcloud import WordCloud
```

## selenium을 이용해 웹 자동화를 위한 webdriver 연동

```python
#크롬 세션 열기
chromedriver = './chromedriver.exe'
driver = webdriver.Chrome(chromedriver) # 빈 브라우저 생성
```

## 크롤링할 웹 페이지 접속

```python
driver.get('https://www.google.com/')
```

## F12 버튼 눌러서 찾은 xpath 를 통해 텍스트 입력

```python
# 검색버튼에 대한 xpath 입력하기
search_bar = driver.find_element_by_xpath('//*[@id="tsf"]/div[2]/div[1]/div[1]/div/div[2]/input')

# 해당 경로에 키워드 입력하기
search_bar.send_keys('단국대학교')
```

## Enter 키를 눌러 검색

```python
search_bar.submit()
```

## webdriver는 켜진 창의 크기에 의존!

#### 현재 켜져있는 창 안에서 누르고자 하는 버튼이 존재하지 않는다면 webdriver도 누르지 못함.

#### 그래서 원하는 버튼이 있는 곳 까지 스크롤을 내려줘야 함.

```python
driver.execute_script("window.scrollTo(0, 700);")
```

## 버튼을 클릭

```python
# 단국대학교 백과사전의 경로를 가져온 후 클릭하기
driver.implicitly_wait(3)# 암묵적으로 웹 자원 로드를 위해 3초까지 기다려 준다.
driver.find_element_by_xpath('/html/body/div[5]/div[2]/div[9]/div[1]/div[2]/div/div[2]/div[2]/div/div/div[3]/div/div[1]/a/h3').click()
```

## Text를 가져오기!

# 네이버 로그인

```python
# 네이버 접속
driver = webdriver.Chrome(chromedriver)
driver.get('https://www.naver.com')
driver.implicitly_wait(3)

# 로그인 버튼 클릭
driver.find_element_by_xpath('/html/body/div[2]/div[3]/div[3]/div/div[2]/a').click()

# 아이디 및 패스워드 입력.
driver.find_element_by_name('id').send_keys('naver_id')
driver.find_element_by_name('pw').send_keys('mypassword1234')

driver.find_element_by_xpath('/html/body/div[2]/div[3]/div/form/fieldset/input').click()
```
